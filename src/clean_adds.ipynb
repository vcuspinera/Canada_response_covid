# Cleaning tweets' database from adds
## Load libraries and open dataset

```python
# General libraries
import pandas as pd
import numpy as np
import time
```

```python
# Parameters
START = time.time()
my_folder = "../tweets/"
```

## Open tweets_db_sentiment.json file

```python
# Open tweets_db_sentiment.json file with all tweets preprocessed and with sentiment scores
df = pd.read_json(my_folder + 'tweets_db_sentiment.json')
```

## `usernames` that most duplicate tweets

```python
pd.set_option('display.max_colwidth', None)
df[df[['account', 'lang','username', 'tweet']]\
   .duplicated()].sort_values(by=['retweetCount', 'likeCount','replyCount','quoteCount'], ascending=False)\
   .groupby(['lang', 'username', 'account'])[['tweet']].count().sort_values(by=['tweet'], ascending=False).head(20)
```

From the previous 25th accounts with most duplication of tweets, we realize the following accounts published annoucements or repeats the same tweet several times:
> "buffydragon", "mikecoulson48", "CigarGuy83", "AmbaPikin", "HaidaPrincess", "andytoronto", "pattena1", "TheCigarDungeon"

## `usernames` that have most tweets with polarity and subjectivity equal to 0

```python
df[(df.polarity==0) & (df.subjectivity==0)].groupby(['username', 'sourceLabel', 'account'])[['tweet']]\
    .count().sort_values(by=['tweet'], ascending=False).head(20)
```

From the previous 25th accounts with with most tweets with polarity and subjectivity equal to 0, none published annoucements or repeated the same tweets several times.
